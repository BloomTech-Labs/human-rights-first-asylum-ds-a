```python
def similar(target_phrases, file):
    
    """GET RID OF PUNCT"""
    
    # create matcher object and add the pattern we are looking for
    matcher = Matcher(nlp.vocab)
    
    matcher.add('target_phrases', target_phrases)
    
    matches = matcher(file, as_spans=True)
    
    return matches
```
