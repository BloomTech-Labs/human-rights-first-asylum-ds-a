```python
import gensim
import re
from english_words import english_words_set
```

```python
with open(r'/content/case_file.txt') as f:
    lines = f.read()
```

```python
x = re.sub(r'\n','',lines)
y = re.sub(r'!','',x)
z = re.sub(r'[0-9]','',y)

def func(string):
  new = []
  for i in string:
    if i.isalpha() or i == ',' or i =='.' or  i == ' ':
      new.append(i)
  return ''.join(new)

new = func(z)
new
```

```python
s = new.split()

def clean(s):
  new = []
  for i in s:
    if i in english_words_set:
      new.append(i)
    elif i == ',' or i == '.':
      new.append(i)
  return ' '.join(new)

fin = clean(s)
fin
```

```python
def summarize(text_data):
  return gensim.summarization.summarize(fin, ratio=0.05)

summarized = summarize(fin)

summarized
```
